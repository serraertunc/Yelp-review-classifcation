## NLP Classification Task

Let's consider the reviews that are either 1 star or 5 stars.

```python
from nltk.corpus import stopwords
```

```python
yelp_class=yelp[(yelp["stars"]== 1) | (yelp["stars"]==5)]
yelp_class.head()
```

Create two objects X and y. (features and target/labels)

```python
X=yelp_class["text"]
y=yelp_class["stars"]
```

Preprocessing function

```python
def text_process(text):
    nopunc = [char for char in text if char not in string.punctuation+string.digits]
    nopunc="".join(nopunc).lower()
    clean=[word for word in nopunc.split() if word not in stopwords.words("english")]
    return " ".join(clean)
```

```python
text_process(" Rosie, Dakota, and I LOVE 56 Chaparral 4 Dog Park!!")
```

## Prediction by using only CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
import string
from nltk.corpus import stopwords
```

```python
X=X.apply(text_process)
```

```python
cv=CountVectorizer(preprocessor=text_process)
X=cv.fit_transform(X)
```

```python
X.shape
```

```python
X.toarray()
```

```python
cv.get_feature_names()
```

Split data into training and testing data.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30,random_state=101)
```

Train a model (MultinomialNB)

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model_CV = MultinomialNB().fit(X_train, y_train)
```

## Predictions and Evaluations by only CountVectorizer

```python
pred_CV=model_CV.predict(X_test)
```

```python
pred_CV
```

Confusion matrix and classification report

```python
from sklearn import metrics
print(metrics.confusion_matrix(y_test,pred_CV))
print(metrics.classification_report(y_test,pred_CV))
```

## Prediction by using CountVectorizer and TF-IDF

```python
from sklearn.feature_extraction.text import TfidfTransformer
```

```python
tfidf_transformer = TfidfTransformer().fit(X)
tfidf_3=tfidf_transformer.transform(X_train[2])
```

```python
print(tfidf_3)
print(tfidf_3.shape)
```

```python
tfidf_X_train=tfidf_transformer.transform(X_train)
tfidf_X_test=tfidf_transformer.transform(X_test)
```

```python
print(tfidf_X_train)
print(tfidf_X_train.shape)
```

```python
model_cv_tfidf = MultinomialNB().fit(tfidf_X_train, y_train)
pred_cv_tfidf=model_cv_tfidf.predict(tfidf_X_test)
```

```python
print(metrics.confusion_matrix(y_test,pred_cv_tfidf))
print(metrics.classification_report(y_test,pred_cv_tfidf))
```

### A pipeline can also be another way to do these steps
pipeline with the following steps:CountVectorizer(), TfidfTransformer(),MultinomialNB()

```python
from sklearn.pipeline import Pipeline
```

```python
pipeline = Pipeline([
    ('bow', CountVectorizer(preprocessor=text_process)),  # strings to token integer counts
    ('tfidf', TfidfTransformer()),  # integer counts to weighted TF-IDF scores
    ('classifier', MultinomialNB()),  # train on TF-IDF vectors w/ Naive Bayes classifier
])
```

Redo the train test split on the yelp_class object.

```python
X = yelp_class['text']
y = yelp_class['stars']
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.3,random_state=101)
```

```python
pipeline.fit(X_train,y_train)
```

### Predictions and Evaluation 

** Now use the pipeline to predict from the X_test and create a classification report and confusion matrix. You should notice strange results.**

```python
predictions= pipeline.predict(X_test)
```

```python
predictions
```

```python
print("Without Pipeline:")
print(metrics.confusion_matrix(y_test,pred_cv_tfidf))
print(metrics.classification_report(y_test,pred_cv_tfidf))
print("With Pipeline:")
print(metrics.confusion_matrix(y_test,predictions))
print(metrics.classification_report(y_test,predictions))
```
